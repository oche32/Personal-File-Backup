<font size="5">Part 1</font> : Preparation and Testing

```python
import pandas as pd
import requests
from scrapy import Selector
import pandas as pd
from pprint import pprint
from tqdm.notebook import tqdm
```

After importing modules, I read my file. Then extracting the values from the "link" and "derived_link" columns in the CSV file, I define this as link_to_use_1 and link_to_use_2 separately.

```python
file = file_path = r'D:\ds105a-2023-w07-summative-oche32\initial data\brief_wikipedia_data.csv'
df = pd.read_csv(file)

link_to_use =[]
link_to_use_1 = df['link'].tolist() 
link_to_use_2 = df['derived_link'].tolist()
```

Then I remove duplicate terms in derived_link

```python
non_duplicate_list2 = list(set(link_to_use_2))

link_to_use.extend(link_to_use_1)
link_to_use.extend(non_duplicate_list2)
```

1: Print the list to verify the extracted links. 2: Check how many links I got

```python
pd.Series(link_to_use)
```

```python
len(link_to_use)
```

Make the link complete

```python
base_url = 'https://en.wikipedia.org'
final_urls = [] 
for link in link_to_use:
    # Converting float values to strings
    if isinstance(link, float):
        link = str(link)
    final_url = base_url + link
    final_urls.append(final_url)
```

Test connections by randomly choosing a data range to test for time-saving

```python
for url in final_urls[1032:1038]:
    response = requests.get(url)
    print(f'URL: {url}, Status Code: {response.status_code}')
```

<font size="5">Part 2</font> : further data

Now it's time to find some interesting data to analyze.  
To start with, I would like to know some information that may contribute my analysis.

**2.1** "last edited time" for each web page.

```python
all_time_data= []
for i in final_urls[0:1379]:
    headers = {'User-Agent': 'Web scraping practice  (c.feng6@lse.ac.uk)'}
    response = requests.get(i, headers = headers)
    sel_time = Selector(response=response)
    time = sel_time.css('li#footer-info-lastmod ::text').getall()
    all_time_data.append(time)
pprint(all_time_data)
```

Save Data

```python
df = pd.DataFrame({'edit_time': all_time_data})
df.to_csv(r'D:\ds105a-2023-w07-summative-oche32\further data\all_time_data.csv', index=False)
```

**2.2** I am also interested in the "stub"(short articles that need further expanding)

```python
all_stub_data= []
for i in final_urls[0:1379]:
    headers = {'User-Agent': 'Web scraping practice  (c.feng6@lse.ac.uk)'}
    response2 = requests.get(i, headers = headers)
    sel_stub = Selector(response = response2)
    stub = sel_stub.css('p.asbox-body ::text').getall()
 # I don't want to get the empty brackets inside my list since not all articles are stubs, so I process the data here.
    if stub:
        all_stub_data.append(stub)
pd.Series(all_stub_data)
```

save data

```python
df = pd.DataFrame({'stubs': all_stub_data})
df.to_csv(r'D:\ds105a-2023-w07-summative-oche32\further data\all_stubs_data.csv', index=False)
```
